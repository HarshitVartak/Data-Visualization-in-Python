#  Data Visualization Python

Python offers several libraries for data visualization, each with its strengths and use cases. Some of the popular libraries for data visualization in Python include Matplotlib & Seaborn

# Matplotlib

Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Plot

A basic plot using Matplotlib typically involves creating an array of data points for the x-axis and another array for the corresponding y-axis. 

```python
x= np.arange(0,10,0.2) # evenly sampled time at 0.2 interval
x
```

```python
y=np.cos(x)
y
```

## Line plot

A line plot (or line chart) is a type of data visualization that displays data points called "markers" connected by straight line segments. 

```python
plt.plot(x)
#plt.axis([0,10,0,1])
plt.xlabel('Time')
plt.ylabel('Tan wave')
plt.title('Line plot')
plt.show()
```

## Multiple plots on same line formating

```python
t = np.arange(0,5,0.2) # evenly sampled time at 0.2 interval
# red dashes , blue squares and green triangles
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^',t,t**4,'y*',linewidth=5.0)
plt.show()
```

## Subplot

Subplots in Matplotlib allow you to create multiple plots in a single figure.

```python
plt.figure(figsize=(20,5))# this argument means the figure will have a width of 20 inches and a height of 5 inches.
 
# The first value 1 indicates the number of rows in the grid of subplots.
# The second value 3 indicates the number of columns in the grid of subplots.
# The third value 1 indicates the position of the specific subplot within this grid.

plt.subplot(1,4,1)
plt.plot(t,t,'r-.')

plt.subplot(1,4,2)
plt.plot(t,t**2,'b:')

plt.subplot(1,4,3)
plt.plot(t,t**3,'y^')

plt.subplot(1,4,4)
plt.plot(t,t**4,'g*')
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
plt.plot(t,t,'r-.')

plt.subplot(2,4,2)
plt.plot(t,t**2,'b:')

plt.subplot(3,4,3)
plt.plot(t,t**3,'y^')

plt.subplot(4,4,4)
plt.plot(t,t**4,'g*')
```

## Scatter Graph

```python
plt.figure(figsize=(7,5))
x=np.random.randn(100)
y=np.random.randn(100)
plt.scatter(x,y)
plt.xlabel('Time')
plt.ylabel('Tan wave')
plt.title('Scatter graph')
```

## Histogram


A histogram is a graphical representation of the distribution of a dataset, it is a type of bar plot

```python
plt.figure(figsize=(7,5))
plt.hist(x, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

## Pie chart

A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [10, 20, 30, 40]
colors = ['yellow', 'green', 'cyan', 'coral']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Pie chart')
plt.show()
```

# Seaborn

* Seaborn is a visualization library that sits on top of matplotlib
* Seaborn offers enhanced features compared to matplotlib

https://seaborn.pydata.org/examples/index.html

```python
import numpy as np               # data manipulation using dataframes
import pandas as pd              # data statical analysis
import matplotlib.pyplot as plt  # data visualization
import seaborn as sns            # statical data visualization
```

```python
# Import data

titanic_data = sns.load_dataset('titanic')
titanic_data.head()
```

```python
titanic_data.columns
```

## Count plot

A count plot is a simple way to visualize the distribution of a categorical variable. 

```python
plt.figure(figsize=(7,5))
sns.countplot(x='class',data=titanic_data)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Count plot')
plt.show()
```

## Bar plot

A bar plot is a common way to represent categorical data using rectangular bars. 

```python
plt.figure(figsize=(7,5))
sns.countplot(x='survived',data=titanic_data)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Bar plot')
plt.show()
```

## Box plot

A box plot (or box-and-whisker plot) is a way to visualize the distribution of a dataset. 

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='class',y='age',data=titanic_data)
plt.xlabel('Class')
plt.ylabel('Age')
plt.title('Box plot')
plt.show()
```

## Swarm plot

A swarm plot is a categorical scatter plot where points do not overlap, providing a better visualization of the distribution of data.

```python
plt.figure(figsize=(7,5))
sns.swarmplot(x='class',y='age',data=titanic_data)
plt.xlabel('Class')
plt.ylabel('Age')
plt.title('Swarm plot')
plt.show()
```

## Joint plot

A joint plot in Seaborn is a combination of univariate and bivariate plots to visualize the relationship between two variables

```python
plt.figure(figsize=(7,5))
sns.jointplot(x='fare',y='age',data=titanic_data)
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Joint plot')
plt.show()
```

## Scatter plot

A scatter plot is a way to visualize the relationship between two continuous variables. 

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='fare',y='age',hue='class',data=titanic_data)
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Scatter plot')
plt.show()
```

## Dist plot

A distribution plot is used to visualize the univariate distribution of a single variable

```python
plt.figure(figsize=(7,5))
sns.distplot(titanic_data['fare'],bins=25,kde=True,color='blue')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Distribution plot')
plt.show()
```

## Heatmap

A heatmap is a graphical representation of data where values in a matrix are represented as colors. 

```python
plt.figure(figsize=(10,10))
sns.heatmap(titanic_data.corr(),annot=True,vmax=1,vmin=-1)
plt.xlabel('Column names')
plt.ylabel('Column names')
plt.title('Heat map')
plt.show()
```

## Pair plot

A pair plot, also known as a scatterplot matrix, is a grid of scatterplots that shows the relationships between pairs of variables in a dataset

```python
plt.figure(figsize=(10,10))
sns.pairplot(titanic_data, vars = ['survived', 'age', 'fare', 'class'])
plt.xlabel('Column names')
plt.ylabel('Column names')
plt.title('Pair plot')
plt.show()
```
